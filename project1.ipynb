```python
#1)	Load in the appropriate csv file as a pandas dataframe (batting.csv)
df = pd.read_csv('Batting.csv')
df
```

```python
#3)	How many players have hit 40 or more HRs in one single season? (Number only)
newdf = df[df['HR']>40]
#
newdf.playerID.nunique()
```

```python
#4)How many players have hit more than 600 HRs for their career? (Dataframe)
#create newdf and groupby unique playerID and sum 'H'
newdf = df.groupby(['playerID','nameFirst','nameLast'])['H'].sum()
#filter out H > 600 and transfer to dataframe
HR600 = newdf[newdf > 600].to_frame(name=None)
#showing result
HR600
```

```python
#5)How many players have hit 40 2Bs, 10 3Bs, 200 Hits, and 30 HRs (inclusive) in one season?(Number Only)
#create newdf as meeting all the criteria
newdf = df[(df['2B']>=40) &(df['3B']>=10) & (df['H']>=200) & (df['HR']>=30)]
#get the unique playerID number
newdf.playerID.nunique()
```

```python
#6)How many players have had 100 or more SBs in a season? (Dataframe)
#create newdf and groupby playerID and yearID, then sum the SB values(avoid situation that player will switch to another team in a season)
newdf = df.groupby(['playerID','nameFirst','nameLast','yearID'])[['SB']].sum()
#create SB100 as meeting 100 or more SBs
SB100 = newdf[newdf['SB']>=100]
#showing result
SB100
```

```python
#7)How many players in the 1960s have hit more than 200 HRs? (Dataframe)
#create newdf as meeting yearID criteria
newdf = df[(df['yearID']>=1960) & (df['yearID']<1970)] 
#create HR200 and groupby playerID then sum all 'HR'values 
HR200 = newdf.groupby(['playerID','nameFirst','nameLast'])[['HR']].sum()
#filter out 'HR' more than 200
HR200[HR200['HR']>200]
```

```python
#8)Who has hit the most HRs in history? (Dataframe)
#create HRmax and groupby playerID and gathering the 'HR' values
HRmax = df.groupby(['playerID','nameFirst','nameLast'])[['HR']].sum()
#show the maximum HR player
HRmax.loc[[HRmax['HR'].idxmax()]]
```

```python
#9)Who had the most hits in the 1970s? (Dataframe)
#create DF70 represent year 1970 to 1979
DF70 = df[(df['yearID']>=1970) & (df['yearID']<1980) ]
#mosthit1970 means groupby the playerID then sum the H values
mosthit1970 = DF70.groupby(['playerID','nameFirst','nameLast'])[['H']].sum()
#showing result
mosthit1970.loc[[mosthit1970['H'].idxmax()]]
```

```python
#10)Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe)
#create a new column OBP = (H+BB+IBB+SH+SF)/AB
df['OBP'] = (df['H']+df['BB']+df['IBB']+df['SH']+df['SF'])/df['AB']
#fill the null values with 0
df['OBP'].fillna(value=0)
#create a new column PA = AB+BB+IBB+SH+SF
df['PA'] = df['AB']+df['BB']+df['IBB']+df['SH']+df['SF']
#fill the null values with 0
df['PA'].fillna(value=0)
#select data in 1977 then create df1977 
df1977 = df[df['yearID']==1977]
#create a dataframe called PA500
PA500 = df1977[df1977['PA']>=500]
#sort by OBP and show the top 5 highest OBP
PA500.sort_values(by='OBP').tail(5)
```

```python
#11)Top 8 highest averages (H/AB)in 2013 with at least 300 PAs? (Dataframe)
#create new column called averages = H/AB
df['average'] = df['H']/df['AB']
#select data in 2013 and at least 300 PAs and called it newdf
newdf = df[(df['yearID']==2013) & (df['PA']>=300)]
#sort newdf to get 8 highest averages
newdf.sort_values(by='average').tail(8)
```

```python
#12)Leaders in hits from 1940 up to and including 1949. (Dataframe)  top 10 and sort by hit
#data from 1940 to 1949
df[(df['yearID']>=1940) & (df['yearID']<=1949)]
#sum up all hits values and call it sumhits
sumhits = df.groupby(['playerID','nameFirst','nameLast'])[['H']].sum()
#sort by hit and select top 10
sumhits.sort_values(by='H').tail(10)
```

```python
#13)Who led MLB (entire data set)with the most hits the most times?  And how many times?  (Dataframe, Number)  
#create newdf from groupby the playerID and sum all the hits values
newdf = df.groupby(['playerID','nameFirst','nameLast'])[['H']].sum()
#sort the newdf by hits and make it as decending order
newdf.sort_values(by='H',ascending=False)
```

```python
#14)Which players have played the most games for their careers?  Top 5, descending by games played presented as a dataframe
#sum the games and groupby playerID with calling it newdf
newdf = df.groupby(['playerID','nameFirst','nameLast'])['G','nameFirst','nameLast'].sum()
#sort it by G and get the top 5 values
newdf.sort_values(by='G',ascending=False).head(5)
```

```python
#15) How many players have had more than 3000 hits for their careers while also hitting 500 or more HRs?  Just a number is okay here
#create newdf represent the sum of each player's hits and homerun times
newdf = df.groupby(['playerID','nameFirst','nameLast'])['H','HR'].sum()
#filter out the data from creating the criteria
newdf[(newdf['H']>3000) & (newdf['HR']>=500)]
```

```python
#16) How many HRs were hit during the entire 1988 season?  Just a number is okay here
#select data in 1988 and called it df1988
df1988 = df[df['yearID']==1988]
#sum each row value of HR column
df1988['HR'].sum()
```

```python
#17) Please filter out and show me the top 3 average seasons by Wade Boggs during his career in seasons in which he had at least 500 ABs.  I would like a dataframe sorted by average.
#create WB means we want only Wade Boggs's data
WB = df[df['nameFirst'].str.contains("Wade") & df['nameLast'].str.contains('Boggs')]
#avoid WB transer to another team in the same year, we groupby yearID and sum AB values
WB.groupby(['yearID'])[['AB']].sum()
#create WB500 representing AB is more than 500
WB500 = WB[WB['AB']>=500]
#create new column 'average'
WB500['average'] = WB500['H'] / WB500['AB']
#sort by 'average' to get the top 3 values
WB500.sort_values(by='average').tail(3)
```

```python
#18) Please filter out the top OBPs for the 1995 season with at least 400 PAs, sorted by OBP.  I would like a dataframe for this
#create a new column OBP = (H+BB+IBB+SH+SF)/AB
df['OBP'] = (df['H']+df['BB']+df['IBB']+df['SH']+df['SF'])/df['AB']
#fill the null values with 0
df['OBP'].fillna(value=0)
#create a new column PA = AB+BB+IBB+SH+SF
df['PA'] = df['AB']+df['BB']+df['IBB']+df['SH']+df['SF']
#fill the null values with 0
df['PA'].fillna(value=0)
#select data in 1995 then create df1995 
df1995 = df[df['yearID']==1995]
#create a dataframe called PA400
PA400 = df1995[df1995['PA']>=400]
#sort by OBP and show the top highest OBP
PA500.sort_values(by='OBP').tail(1)
```

```python
#19) Who had the most 3Bs (in total) in 1922, 1925, 1926, and 1928?  I would like a dataframe with just the leader
#In df1922 we find the top total 3Bs player
df1922 = df[(df['yearID']==1922)] 
sumdf1922 = df1922.groupby(['playerID','nameFirst','nameLast'])[['3B','yearID']].sum()
max3b_1922 = sumdf1922.sort_values(by='3B').tail(1)

#In df1925 we find the top total 3Bs player
df1925 = df[(df['yearID']==1925)]
sumdf1925 = df1925.groupby(['playerID','nameFirst','nameLast'])[['3B','yearID']].sum()
max3b_1925 = sumdf1925.sort_values(by='3B').tail(1)

#In df1926 we find the top total 3Bs player
df1926 = df[(df['yearID']==1926)]
sumdf1926 = df1926.groupby(['playerID','nameFirst','nameLast'])[['3B','yearID']].sum()
max3b_1926 =  sumdf1926.sort_values(by='3B').tail(1)

#In df1928 we find the top total 3Bs player
df1928 = df[(df['yearID']==1928)]
sumdf1928 = df1928.groupby(['playerID','nameFirst','nameLast'])[['3B','yearID']].sum()
max3b_1928 = sumdf1928.sort_values(by='3B').tail(1)

#put all the data in the same dataframe
pd.concat([max3b_1922,max3b_1925,max3b_1926,max3b_1928])
```

```python
#22) Same question but for 1997 and only in the NL (check league ID)? (Dataframe)
#create a new column OBP = (H+BB+IBB+SH+SF)/AB
df['OBP'] = (df['H']+df['BB']+df['IBB']+df['SH']+df['SF'])/df['AB']
#fill the null values with 0
df['OBP'].fillna(value=0)
#create a new column PA = AB+BB+IBB+SH+SF
df['PA'] = df['AB']+df['BB']+df['IBB']+df['SH']+df['SF']
#fill the null values with 0
df['PA'].fillna(value=0)
#select data in 1997 then create df1997 
df1997 = df[df['yearID']==1997]
#create a dataframe called PA400 and include criteria PA = 400 and league ID = NL
PA400 = df1997[(df1997['PA']>=400) & (df1997['lgID']=='NL')]
#sort by OBP and show the top highest OBP
PA400.sort_values(by='OBP').tail(1)
```

```python
#23) Who had more than the league average HRs in 2012 (filter out all players with less 500 PAs)? (Dataframe)
#create a new column PA = AB+BB+IBB+SH+SF
df['PA'] = df['AB']+df['BB']+df['IBB']+df['SH']+df['SF']
#fill the null values with 0
df['PA'].fillna(value=0)
#select data in 2012 then create df2012 
df2012 = df[df['yearID']==2012]
#calculate average HR in 2012
average_HR_2012 = df2012.HR.mean()
#create a dataframe called PAless500 and include criteria PA < 500 
PAless500 = df2012[(df2012['PA']<500) & (df2012['HR']>average_HR_2012)]
#show outcome
PAless500
```
