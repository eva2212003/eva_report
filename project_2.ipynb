 # Import data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import cufflinks as cf
```

```python
#read each csv file
AAPL = pd.read_csv('AAPL_2006-01-01_to_2018-01-01.csv')
AMZN = pd.read_csv('AMZN_2006-01-01_to_2018-01-01.csv')
CSCO = pd.read_csv('CSCO_2006-01-01_to_2018-01-01.csv')
GOOGL = pd.read_csv('GOOGL_2006-01-01_to_2018-01-01.csv')
IBM = pd.read_csv('IBM_2006-01-01_to_2018-01-01.csv')
VZ = pd.read_csv('VZ_2006-01-01_to_2018-01-01.csv')
```

# Basic Analyses

```python
AAPL = AAPL.set_index('Date')
AMZN = AMZN.set_index('Date')
GOOGL = GOOGL.set_index('Date')
IBM = IBM.set_index('Date')
VZ = VZ.set_index('Date')
CSCO = CSCO.set_index('Date')
```

```python
#set new variable of each close price
AAPL_close = AAPL[['Close']]
AMZN_close = AMZN[['Close']]
GOOGL_close = GOOGL[['Close']]
IBM_close = IBM[['Close']]
VZ_close = VZ[['Close']]
CSCO_close = CSCO[['Close']]
```

```python
#set the figure size
plt.figure(figsize=(20,10))

#plot each stock price by date
plt.plot(AAPL_close, color='red', label='APPLE')
plt.plot(AMZN_close, color='green', label='AMAZON')
plt.plot(GOOGL_close, color='blue', label='GOOGLE')
plt.plot(IBM_close, color='orange', label='IBM')
plt.plot(VZ_close, color='purple', label='VERIZON')
plt.plot(CSCO_close, color='black', label='CISCO')


#set the labels and title
plt.xlabel('date')
plt.ylabel('price')
plt.title('comparison of stock price among 6 companies')
plt.legend()
```

As we can see in the plot, Amazon's stock price is the highest recently among the 6 companies.
I would like to discuss Amazon and Google, the two highest stock price companies.

# Compare the feature (Google vs. Amazon)

We would like to get a closer look to the comparison od Google and Amazon

When I compare the price difference of each day, Amazon maintains stable price difference for most of the time, but in 2014 the range of the price of the same day has begun dramatically larger than before. Unlike Amazon, Google always have a midium-large range price difference in the same day.
